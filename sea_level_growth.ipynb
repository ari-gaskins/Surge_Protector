```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# display all available columns (can do with rows)
pd.set_option("display.max_columns", 28)
```

```python
sea_levels = pd.read_csv('sea_level_projections.csv')
```

```python
filter_pt3_low = (sea_levels['Scenario'] == '0.3 - LOW')
filter_pt3_med = (sea_levels['Scenario'] == '0.3 - MED')
filter_pt3_high = (sea_levels['Scenario'] == '0.3 - HIGH')

filter_pt5_low = (sea_levels['Scenario'] == '0.5 - LOW')
filter_pt5_med = (sea_levels['Scenario'] == '0.5 - MED')
filter_pt5_high = (sea_levels['Scenario'] == '0.5 - HIGH')

filter_1pt_low = (sea_levels['Scenario'] == '1.0 - LOW')
filter_1pt_med = (sea_levels['Scenario'] == '1.0 - MED')
filter_1pt_high = (sea_levels['Scenario'] == '1.0 - HIGH')

filter_1pt5_low = (sea_levels['Scenario'] == '1.5 - LOW')
filter_1pt5_med = (sea_levels['Scenario'] == '1.5 - MED')
filter_1pt5_high = (sea_levels['Scenario'] == '1.5 - HIGH')

filter_2pt_low = (sea_levels['Scenario'] == '2.0 - LOW')
filter_2pt_med = (sea_levels['Scenario'] == '2.0 - MED')
filter_2pt_high = (sea_levels['Scenario'] == '2.0 - HIGH')
```

```python
sea_levels_pt3_low = sea_levels.loc[filter_pt3_low]
sea_levels_pt3_med = sea_levels.loc[filter_pt3_med]
sea_levels_pt3_high = sea_levels.loc[filter_pt3_high]

sea_levels_pt5_low = sea_levels.loc[filter_pt5_low]
sea_levels_pt5_med = sea_levels.loc[filter_pt5_med]
sea_levels_pt5_high = sea_levels.loc[filter_pt5_high]

sea_levels_1pt_low = sea_levels.loc[filter_1pt_low]
sea_levels_1pt_med = sea_levels.loc[filter_1pt_med]
sea_levels_1pt_high = sea_levels.loc[filter_1pt_high]

sea_levels_1pt5_low = sea_levels.loc[filter_1pt5_low]
sea_levels_1pt5_med = sea_levels.loc[filter_1pt5_med]
sea_levels_1pt5_high = sea_levels.loc[filter_1pt5_high]

sea_levels_2pt_low = sea_levels.loc[filter_2pt_low]
sea_levels_2pt_med = sea_levels.loc[filter_2pt_med]
sea_levels_2pt_high = sea_levels.loc[filter_2pt_high]
```

```python
years = ['RSL2020 (cm)', 'RSL2030 (cm)', 'RSL2040 (cm)', 'RSL2050 (cm)',
                    'RSL2060 (cm)', 'RSL2070 (cm)', 'RSL2080 (cm)', 'RSL2090 (cm)', 
                     'RSL2100 (cm)', 'RSL2110 (cm)', 'RSL2120 (cm)', 'RSL2130 (cm)',
                    'RSL2140 (cm)', 'RSL2150 (cm)']
num_years = [2020, 2030, 2040, 2050, 2060, 2070, 2080, 2090, 2100, 2110, 2120, 2130,
            2140, 2150]
```

```python
pt3_low_avgs = sea_levels_pt3_low[years].mean()
pt3_med_avgs = sea_levels_pt3_med[years].mean()
pt3_high_avgs = sea_levels_pt3_high[years].mean()

pt5_low_avgs = sea_levels_pt5_low[years].mean()
pt5_med_avgs = sea_levels_pt5_med[years].mean()
pt5_high_avgs = sea_levels_pt5_high[years].mean()

_1pt_low_avgs = sea_levels_1pt_low[years].mean()
_1pt_med_avgs = sea_levels_1pt_med[years].mean()
_1pt_high_avgs = sea_levels_1pt_high[years].mean()

_1pt5_low_avgs = sea_levels_1pt5_low[years].mean()
_1pt5_med_avgs = sea_levels_1pt5_med[years].mean()
_1pt5_high_avgs = sea_levels_1pt5_high[years].mean()

_2pt_low_avgs = sea_levels_2pt_low[years].mean()
_2pt_med_avgs = sea_levels_2pt_med[years].mean()
_2pt_high_avgs = sea_levels_2pt_high[years].mean()
```

```python
avg_sea_level_rise = pd.DataFrame({
    '0.3 m - LOW': pt3_low_avgs,
    '0.3 m - MED': pt3_med_avgs, 
    '0.3 m - HIGH': pt3_high_avgs,
    '0.5 m - LOW': pt5_low_avgs, 
    '0.5 m - MED': pt5_med_avgs, 
    '0.5 m - HIGH': pt5_high_avgs,
    '1.0 m - LOW': _1pt_low_avgs, 
    '1.0 m - MED': _1pt_med_avgs, 
    '1.0 m - HIGH': _1pt_high_avgs,
    '1.5 m - LOW': _1pt5_low_avgs, 
    '1.5 m - MED': _1pt5_med_avgs, 
    '1.5 m - HIGH': _1pt5_high_avgs,
    '2.0 m - LOW': _2pt_low_avgs, 
    '2.0 m - MED': _2pt_med_avgs, 
    '2.0 m - HIGH': _2pt_high_avgs
})
```

```python
avg_sea_level_rise
```

```python
avg_sea_level_rise.plot(kind='bar')
```

```python
avg_sea_level_rise.head().plot(kind='bar')
```

```python
pt3_avg = (pt3_low_avgs + pt3_med_avgs + pt3_high_avgs) / 3
pt5_avg = (pt5_low_avgs + pt5_med_avgs + pt5_high_avgs) / 3
_1pt_avg = (_1pt_low_avgs + _1pt_med_avgs + _1pt_high_avgs) / 3
_1pt5_avg = (_1pt5_low_avgs + _1pt5_med_avgs + _1pt5_high_avgs) / 3
_2pt_avg = (_2pt_low_avgs + _2pt_med_avgs + _2pt_high_avgs) / 3
```

```python
avg_sea_level_rise = pd.DataFrame({
'0.3 m': pt3_avg, 
'0.5 m': pt5_avg, 
'1.0 m': _1pt_avg, 
'1.5 m': _1pt5_avg, 
'2.0 m': _2pt_avg
})
```

```python
avg_sea_level_rise.plot(kind='bar')
```
